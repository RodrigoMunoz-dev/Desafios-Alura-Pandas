```python
from datetime import datetime

datos_locales['datas_combinadas_pagamento'] = datos_locales['datas_combinadas_pagamento'].apply(
    lambda fechas: [datetime.strptime(fecha, '%d/%m/%Y') for fecha in fechas]
)
```

```python
from datetime import datetime

def convertir_lista(fechas):
    nuevas = []
    for f in fechas:
        if isinstance(f, str):
            nuevas.append(datetime.strptime(f, '%d/%m/%Y'))
        else:
            nuevas.append(f)  # ya es datetime
    return nuevas

datos_locales['datas_combinadas_pagamento'] = datos_locales['datas_combinadas_pagamento'].apply(convertir_lista)
```
